```python
#Aaustin crime
client = bigquery.Client() #start big query client

QUERY = (
    'SELECT * FROM  `bigquery-public-data.austin_crime.crime` '
    'ORDER BY primary_type DESC,year DESC,district DESC')

query_job = client.query(QUERY)    # Start Query API Request
query_result = query_job.result()  # Get Query Result
crime_Austin = query_result.to_dataframe()  #save the query result to data frame
```

```python
class Analyze_Crime:
    
    def __init__ (self, dataset):
        
        self.dataset = dataset
        
    
    def crimetype (self,year): #shows number of crimes per given year
        
        data_cut = self.dataset[(self.dataset['year'] == year)]
        print(data_cut['primary_type'].value_counts(sort=True))
        
```
